```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
```

```python
with open("risco_credito.pkl","rb") as f:
    x_risco_credito,y_risco_credito = pickle.load(f)
```

```python
x_risco_credito = np.delete(x_risco_credito, [2,7,11],axis = 0)
y_risco_credito = np.delete(y_risco_credito, [2,7,11],axis = 0)
```

```python
from sklearn.linear_model import LogisticRegression
risco_credito_logistic = LogisticRegression(random_state = 1)
risco_credito_logistic.fit(x_risco_credito,y_risco_credito)
```

```python
risco_credito_logistic.intercept_
```

```python
risco_credito_logistic.coef_
```

```python
previsoes1 = risco_credito_logistic.predict([[0,0,1,2],[2,0,0,0]])
previsoes1
```

```python
with open("credits.pkl","rb") as f:
    x_credit_treinamento,x_credit_teste,y_credit_treinamento,y_credit_teste = pickle.load(f)
```

```python
logistic_credits = LogisticRegression(random_state=1)
logistic_credits.fit(x_credit_treinamento,y_credit_treinamento)
```

```python
logistic_credits.intercept_
```

```python
logistic_credits.coef_
```

```python
previsoes2 = logistic_credits.predict(x_credit_teste)
previsoes2
```

```python
logistic_credits.score(x_credit_teste,y_credit_teste)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_credit_teste,previsoes2))
```
