```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle

base_cesus = pd.read_csv("credit_data.csv")
```

```python
from sklearn.tree import DecisionTreeClassifier
with open("credits.pkl","rb") as f:
    x_credit_treinamento,x_credit_teste,y_credit_treinamento,y_credit_teste = pickle.load(f)
```

```python
arvore_credit = DecisionTreeClassifier(criterion="entropy",random_state=0)
arvore_credit.fit(x_credit_treinamento,y_credit_treinamento)
```

```python
previsoes = arvore_credit.predict(x_credit_teste)
previsoes
```

```python
arvore_credit.score(x_credit_teste,y_credit_teste)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_credit_teste,previsoes))
```

```python
from sklearn import tree
previsoes = ["income","age","loan"]
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(20,20))
tree.plot_tree(arvore_credit,feature_names=previsoes,class_names=["0","1"],filled=True);
```
