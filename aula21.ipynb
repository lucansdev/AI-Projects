```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
```

```python
with open("census.pkl","rb") as f:
    x_census_treinamento,y_census_treinamento,x_census_teste,y_census_teste = pickle.load(f)
```

```python
from sklearn.neural_network import MLPClassifier
rede_neural_census = MLPClassifier(max_iter=1000,verbose=True,hidden_layer_sizes=(100,100),tol=0.0000100)
rede_neural_census.fit(x_census_treinamento,y_census_treinamento)
```

```python
previsoes = rede_neural_census.predict(x_census_teste)
previsoes
```

```python
rede_neural_census.score(x_census_teste,y_census_teste)
```
